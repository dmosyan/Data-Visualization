```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Ploting x and y')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.9,0.9,0.8])

axes.plot(x,y, 'b-')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')

```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.3,0.4])
axes.plot(x,y, 'b-')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')

axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y, 'b--')
axes.set_xlabel('high')
axes.set_ylabel('Weight')
axes.set_title('H vs Weight')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y,'b')
plt.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x,'r')
```

```python
fig = plt.figure(figsize=(5,4))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()

fig.savefig('ploting1.jpg', dpi=200)
```

```python
fig = plt.figure(dpi=100)
ax = fig.add_axes([0,0,0.5,0.5])

ax.plot(x,x**2, label='X squered')
ax.plot(x,x**3, label='X cube')

ax.legend(loc=0)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='purple', linewidth=1, alpha=0.9, linestyle='--',marker='o', markersize=7,
         markerfacecolor='blue', markeredgewidth=5, markeredgecolor='red')

axes.set_xlim(0,1)
axes.set_ylim(0,2)
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,'b-')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,'r')
ax2.plot(x,y,'r')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.plot(x,y)
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.set_title('zoom')
```
